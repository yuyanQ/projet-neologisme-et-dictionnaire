```python
import json
import datetime
from collections import defaultdict

with open('./1999-2003/neo/neo.json', 'r') as f:
    neo_dict = json.load(f)

with open('./1999-2003//neo_emprunt/neo_emprunt.json', 'r') as f:
    neo_emprunt_dict = json.load(f)
```

```python
data_dict = defaultdict(lambda: {'neo_count': 0, 'neo_emprunt_count': 0})
for year, neo_list in neo_dict.items():
    data_dict[year[-4:]]['neo_count'] = len(neo_list)
for year, neo_emprunt_list in neo_emprunt_dict.items():
    data_dict[year[-4:]]['neo_emprunt_count'] = len(neo_emprunt_list)
for year in neo_emprunt_dict.keys():
    data_dict[year[-4:]]['neo_emprunt_pro'] = data_dict[year[-4:]]['neo_emprunt_count']/data_dict[year[-4:]]['neo_count']*100
```

```python
import matplotlib.pyplot as plt

# 
years = sorted(list(data_dict.keys()))
neo_counts = [data_dict[year]['neo_count'] for year in years]
neo_emprunt_counts = [data_dict[year]['neo_emprunt_count'] for year in years]
pro = [data_dict[year]['neo_emprunt_pro'] for year in years]

fig, ax1 = plt.subplots()

# 
ax1.bar(years, neo_counts, width=0.4, label='Nouveau mots')
for i, count in enumerate(neo_counts):
    ax1.text(years[i], count + 50, str(count), ha='center')

ax1.bar(years, neo_emprunt_counts, width=0.4, label='Emprunts')
for i, count in enumerate(neo_emprunt_counts):
    ax1.text(years[i], count + 50, str(count), ha='center')
# 
ax1.set_xlabel('Anne')
ax1.set_ylabel('Nombre de nouveaux mots')
ax1.set_title('Nouveaux mots et emprunts d\'anglais par anne')

# 
ax1.legend()

# 
ax2 = ax1.twinx()

# 
ax2.plot(years, pro, color='red', label="Taux d'emprunts (%)")
for i, p in enumerate(pro):
    ax2.text(years[i], p + 0.02, f'{p:.2f}', ha='center')
ax2.set_ylabel('Taux d\'emprunts')

# y
ax2.legend(loc='upper left')

# 
plt.show()
```
